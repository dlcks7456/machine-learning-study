# Evaluation

### MAE (Mean Absolute Error)
실제 값과 예측 값 차이의 절대값들의 평균  
$MAE = \frac{1}{n}\displaystyle\sum_{i=1}^{n}|y_i-\widehat{y_i}|$  

example)  
$MAE = \frac{10+13+9+19+13}{5} = 12.8$   

### MSE (Mean Squared Error)
실제 값과 예측 값 차이의 제곱한 값들의 평균  

$MAE = \frac{1}{n}\displaystyle\sum_{i=1}^{n}(y_i-\widehat{y_i})^2$  

$MAE = \frac{100+169+81+361+169}{5} = 176$   


**MAE, MSE 모두 음수가 나오는 문제에 대해서 해결을 할 수 있다.**

MAE의 경우 **이상값에 조금 더 민감**하다.  
**단위도 달라질 수 있기 때문에 추가적인 해석이 필요 할 수 있다.**  
(1 미만의 값은 제곱할 때 값이 작아지고, 1 이상의 값은 제곱할 때 값이 커지기 때문)  



###  RMSE (Root Mean Squared Error)
MSE에 루트를 적용

$RMSE = \sqrt{MSE} = \sqrt{\frac{1}{n}\displaystyle\sum_{i=1}^{n}(y_i-\widehat{y_i})^2}$  

$RMAE = \sqrt{\frac{100+169+81+361+169}{5}} = 13.27$   


**MSE가 가지고 있던 제곱에의해 발생한 외곡되는 특성을 방지**할 수 있으며 **단위도 통일**되기 때문에 해석에 용이

### 모두 0에 가까워질 수록 모델 성능이 좋다는 것 (실제값과 예측값이 작을 수록 좋은 것)

### $R^2$ (R square) 결정계수
데이터의 분산을 기반으로 한 평가 지표  

$R^2 = 1 - \frac{SSE}{SST}=\frac{SSR}{SST}$  

$error = 실제값 - 예측값$  
$reg = 예측값 - 평균값$  
$totla = 실제값 - 평균값$  

$total = reg + error$  

각 값에 제곱을한다. (**S**quare)  
$Serror = (실제값 - 예측값)^2$  
$Sreg = (예측값 - 평균값)^2$  
$Stotla = (실제값 - 평균값)^2$  

각 값에 해당하는 모든 y의 값들의 합 (**S**um)  
$SSerror = \sum(실제값 - 예측값)^2$  
$SSreg = \sum(예측값 - 평균값)^2$  
$SStotla = \sum(실제값 - 평균값)^2$  

**S**um of **S**quares for {**E**rror, **R**eg, **T**otal}  

$SSE = \sum(y_i - \widehat{y_i})^2$  
$SSR = \sum(\widehat{y_i} - \overline{y_i})^2$  
$SST = \sum(y_i - \overline{y_i})^2$  


$R^2 = 1 - \frac{SSE}{SST} = \frac{SSR}{SST}$  

$R^2 = 1 - \frac{(실제값 - 예측값)^2의 합}{(실제값 - 평균값)^2의 합}$  

분자의 예측값이 평균값과 비슷하다면, 즉 예측이 평균과 비슷하다면?  
1에 가까워질 것이고 1-1 은 0에 가까워지는 값이 될 것  

예측을 잘해서 실제값과 예측값이 비슷하다면 분자 부분이 0에 가까워질 것이고 값이 0에 가까워지므로 1-0으로 1에 가까워진다.  

즉, $R^2$는 1에 가까워질 수록 모델이 유의하다는 것, 0에 가까워질 수록 모델이 유의하지 않다는 것
