# 4. Logistic Regression (로지스틱 회귀)

지도 학습 중에서 **분류(Classification)**에 해당   
**분류** : **주어진 데이터를 정해진 범주(category)에 따라 분류**  
**예측 결과가 숫자가 아닐 때**   

선형 회귀 방식을 분류에 적용한 알고리즘,  
데이터가 어떤 범주에 속할 확률을 0~1 사이의 값으로 예측,   
더 높은 범주에 속하는 쪽으로 분류  

**범주 : True/False , Yes/No, 합격/불합격 ...**  

example) 스팸 메일, 은행 대출, 악성 여부, 고객의 제품 구매 의사  



### 시그모이드(Sigmoid) 함수 = 로지스틱(Logistic) 함수

**단순 선형** : $y = mx + b$  
**Sigmoide Function** : $P = \frac{1}{1 + e^{-y}}$  
=> $ln\begin{pmatrix}\frac{P}{1-P}\end{pmatrix}= mx + b$  


$0 \leq P(pass) \leq 1$

# 4. Logistic Regression

### 공부 시간에 따른 자격증 시험 합격 가능성

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("LogisticRegressionData.csv")
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### 데이터 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### 학습 (로지스틱 회귀 모델)

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

### 6시간 공부했을 때 예측?

```python
classifier.predict([[6]])
# 결과 1 : 합격할 것으로 예측
```

```python
classifier.predict_proba([[6]]) # 합격할 확률 출력
# 불합격 확률이 14%, 합격할 확류이 86%
```

```python
classifier.predict([[4]])
# 결과 0 : 불합격할 것으로 예측
```

```python
classifier.predict_proba([[4]]) # 합격할 확률 출력
# 불합격 확률이 62%, 합격할 확류이 38%
```

### 분류 결과 예측 (테스트 세트)

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test # 실제 값 (테스트 세트)
```

```python
X_test # 공부 시간 (테스트 세트)
```

```python
classifier.score(X_test, y_test) # 모델 평가
# 전체 테스트 세트 4개 중에서 분류 예측을 올바로 맞힌 갯수 3개 => 3/4 = 0.75
```

### 데이터 시각화 (훈련 세트)

```python
X_range = np.arange(min(X), max(X), 0.1) # X의 최소값에서 최대값까지를 0.1 딘위로 잘라서 데이터 생성
X_range
```

#### Sigmoire function
$P = \frac{1}{1 + e^y}$  
$P = \frac{1}{1 + np.exp(-y)}$  // $y=mx+b$

```python
#p = 1 / (1 + np.exp(-(m * X_range + b)))
p = 1 / (1 + np.exp(-(classifier.coef_ * X_range + classifier.intercept_))) # y = mx + b
p
```

```python
p.shape
```

```python
X_range.shape
```

```python
p = p.reshape(-1) # 1차원 배열 형태로 변경
p.shape
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_range, p, color='green')
plt.plot(X_range, np.full(len(X_range), 0.5), color='red') # X_range의 길이(개수) 만큼 0.5를 채우라는 것
plt.title("Probability by hours")
plt.xlabel('hours')
plt.ylabel('P')
plt.show()
```

### 데이터 시각화 (테스트 세트)

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_range, p, color='green')
plt.plot(X_range, np.full(len(X_range), 0.5), color='red') # X_range의 길이(개수) 만큼 0.5를 채우라는 것
plt.title("Probability by hours")
plt.xlabel('hours')
plt.ylabel('P')
plt.show()
```

```python
classifier.predict_proba([[4.5]]) # 4.5 시간 공부했을 때 확률 (모델에서는 51% 합격로 예측, 실제로는 불합격)
```
