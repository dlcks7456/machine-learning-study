# Gradient Descent (경사 하강법)

실제 값과 예측 값 차이의 제곱의 합을 최소화  
$sum(y - \hat{y})$

- 잔차 제곱의 합 : RSS (Residual Sum of Squares) = SSR (Sum of Squared Residuals)
- 최소제곱법 : OLS (Ordinary Least Squares) = Least Square Method  
최소제곱법은 노이즈에 취약하다. (이상값에 취약)


경사하강법 : 경사를 따라서 하강을 하는 것

$y = mx+b$ 에서  $y = mx$ 를 주목







### 기울기가 0인 지점까지 오는 것이 목표

### 학습률 (Learning rate)
기울기가 0까지 가는 동안 이동하는 비율을 설정해주는 것
- 비율이 낮으면 반복횟수가 많아져서 시간이 오래 걸릴 것
- 비율이 너무 크면 기울기가 0인 곳으로 수렴하지 못할 수 있음 (벗어날 수 있음)

일반적으로는  
0.001, 0.003, 0.01, 0.03, 0.1, 0.3이 많이 사용됨


### 에포크 (Epoch)
경사하강법에서 최적의 값을 찾기 위해 훈련세트에서 모든 데이터들을 한번씩 사용하는 과정

모든 데이터를 사용하는 경우 자원을 많이 소비하게 됨 (시간 소요가 커짐)

이를 보완하기 위한 것이

## Stochastick Gradient Descent (확률적 경사 하강법)
매 단계마다 하나의 데이터를 무작위로 선택하고 그 데이터에 대해서만 기울기를 계산한다.
전체 데이터를 사용할 때까지 사용
